```python
# ============================================================
# M-TriX: The Master Proof (v0.6.1)
# Author: Mohammed M. Alanazi
# Paper (Zenodo): https://zenodo.org/records/17613192
#
# This notebook reproduces the full experimental pipeline used
# in the paper:
#   “M-TriX: A Hybrid-Causal Neural Architecture…”
#
# It implements the exact v0.6.0 design, including:
#   1. Noise-Augmented Training (Robust Training Regime)
#   2. Proof #1: Efficiency–Accuracy Sweep Test
#   3. Proof #2: Noise Robustness Test (Dense vs M-TriX)
#   4. Proof #3: Visual Evidence — Clean p-Map Filtering
#
# This code is the OFFICIAL reproducibility implementation.
# ============================================================

import os
import math
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# -----------------------------
# Global Configuration
# -----------------------------
IMG_SIZE = 32
NUM_CLASSES = 10
BATCH_SIZE = 128
EPOCHS = 10

# Gating configuration
P_MIN = 0.05               # Minimum allowed p-value
BALANCE_TARGET = 0.5       # Target global average p-value
LAMBDA_BAL = 0.02          # Weight for balance regularizer
LAMBDA_L1P = 0.001         # Weight for L1 sparsity regularizer
LAMBDA_SMOOTH = 0.005      # Weight for smoothness regularizer

OVERLAP = True             # Use overlapping downsampling (O(1) stride)
SEED = 1337                # Global seed
NOISE_FACTOR = 0.3         # 30% pixel-level noise

tf.keras.utils.set_random_seed(SEED)
np.random.seed(SEED)


# ---------------------------------------------------------
# 1) MNIST Loading & Noise Augmentation (v0.6.0 Design)
# ---------------------------------------------------------

def add_noise(images, noise_factor=0.3):
    """
    Adds pixel replacement noise:
    - noise_factor defines % of pixels to replace
    - each replaced pixel gets a random value in [0,1]
    """
    noisy_images = images.copy()
    if not noisy_images.flags.writeable:
        noisy_images = np.array(noisy_images, copy=True)

    num_noise_pixels = int(noise_factor * IMG_SIZE * IMG_SIZE)

    for i in range(noisy_images.shape[0]):
        x_coords = np.random.randint(0, IMG_SIZE, num_noise_pixels)
        y_coords = np.random.randint(0, IMG_SIZE, num_noise_pixels)
        values = np.random.rand(num_noise_pixels)
        noisy_images[i, x_coords, y_coords, 0] = values

    return noisy_images


print("Preparing MNIST dataset...")
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize & reshape
x_train = (x_train.astype("float32") / 255.0)[..., None]
x_test  = (x_test.astype("float32")  / 255.0)[..., None]

# Resize to 32×32
x_train = tf.image.resize(x_train, (IMG_SIZE, IMG_SIZE)).numpy()
x_test  = tf.image.resize(x_test,  (IMG_SIZE, IMG_SIZE)).numpy()

# Create noisy training dataset
print(f"Applying {NOISE_FACTOR*100:.0f}% noise to training data...")
x_train_noisy = add_noise(x_train, NOISE_FACTOR)

# Noisy test set (used for Noise Robustness evaluation)
x_test_noisy = add_noise(x_test, NOISE_FACTOR)

test_ds_noisy = tf.data.Dataset.from_tensor_slices((x_test_noisy, y_test))\
    .batch(BATCH_SIZE).prefetch(2)
test_ds_clean = tf.data.Dataset.from_tensor_slices((x_test, y_test))\
    .batch(BATCH_SIZE).prefetch(2)

# Final training dataset: noisy inputs, clean labels
train_ds = tf.data.Dataset.from_tensor_slices((x_train_noisy, y_train))\
    .shuffle(60000, seed=SEED).batch(BATCH_SIZE).prefetch(2)



# ---------------------------------------------------------
# 2) Gated Residual Block (v0.6.0 Robust Formulation)
# ---------------------------------------------------------
class GatedResidualBlock(layers.Layer):
    """
    Implements the v0.6.0 immune design:
       p = G(f(x))
    where the gate operates ONLY on the cleaned, denoised activation f(x).
    """

    def __init__(self, filters, kernel_size=3, stride=1, name_prefix="grb",
                 blackout_switch=None,
                 force_p_value_link=None,
                 **kwargs):
        super().__init__(**kwargs)

        self.filters = filters
        self.stride = stride
        self.name_prefix = name_prefix
        self.blackout_switch = blackout_switch
        self.force_p_value_link = force_p_value_link

        # Smart path f(x)
        self.conv = layers.Conv2D(filters, kernel_size, strides=stride,
                                  padding="same", kernel_initializer="he_normal",
                                  name=f"{name_prefix}_conv")
        self.bn = layers.BatchNormalization(name=f"{name_prefix}_bn")

        # Gating network G(f(x))
        self.gating_net = layers.Conv2D(
            1, kernel_size=1, padding="same", activation="sigmoid",
            name=f"{name_prefix}_gate1x1"
        )

        # Residual projection path
        self.proj = None

        # Smoothness filter (3×3 avg)
        avg_kernel = np.ones((3,3,1,1), dtype=np.float32) / 9.0
        self.avg_filter = tf.constant(avg_kernel, dtype=tf.float32)

        # For visualization
        self.last_p_map_raw = None
        self.last_p_map = None

    def build(self, input_shape):
        in_ch = int(input_shape[-1])
        if in_ch != self.filters or self.stride != 1:
            self.proj = layers.Conv2D(
                self.filters, kernel_size=1, strides=self.stride,
                padding="same", use_bias=False,
                name=f"{self.name_prefix}_proj"
            )
        else:
            self.proj = layers.Lambda(lambda z: z)  # Identity

    def call(self, x, training=None):

        # ----- 1. Compute clean activation f(x)
        y = self.conv(x)
        y = self.bn(y, training=training)
        y_act = tf.nn.relu(y)

        # ----- 2. Compute gating signal p = G(f(x))
        p_raw = self.gating_net(y_act)
        self.last_p_map_raw = p_raw

        # Enforce minimum p to avoid full shutdown
        p_normal = p_raw * (1 - P_MIN) + P_MIN

        # ----- 3. Blackout override (used for Sweep Test)
        def use_forced():
            return tf.ones_like(p_normal) * self.force_p_value_link
        def use_normal():
            return p_normal
        p_final = tf.cond(self.blackout_switch, use_forced, use_normal)

        # ----- 4. Residual projection
        x_proj = self.proj(x)

        # ----- 5. Hybrid blend
        out = y_act * p_final + x_proj * (1 - p_final)

        # ----- Regularization (applies only during training)
        if training:
            # L1 sparsity
            self.add_loss(LAMBDA_L1P * tf.reduce_mean(p_raw))

            # Smoothness penalty
            C = p_raw.shape[-1] or 1
            kernel = tf.tile(self.avg_filter, [1,1,C,1])
            blur = tf.nn.depthwise_conv2d(
                p_raw, kernel, strides=[1,1,1,1], padding="SAME"
            )
            smooth = tf.reduce_mean((p_raw - blur)**2)
            self.add_loss(LAMBDA_SMOOTH * smooth)

            # Balance regularizer
            p_mean = tf.reduce_mean(p_raw)
            self.add_loss(LAMBDA_BAL * (p_mean - BALANCE_TARGET)**2)

        self.last_p_map = p_final
        return out



# ---------------------------------------------------------
# 3) MTMA Model (v0.6.0 Architecture)
# ---------------------------------------------------------
class MTMAModel(keras.Model):
    """
    Full multi-stage MTMA architecture including:
    - Stem
    - 3 Residual Stages with GRBs
    - GAP → Dense head
    """

    def __init__(self, num_classes=10, overlap=True, **kwargs):
        super().__init__(**kwargs)

        self.overlap = overlap
        self.all_grb_layers = []

        # Stem
        self.stem = layers.Conv2D(32, 3, padding="same", activation="relu")
        self.bn0 = layers.BatchNormalization()

        # Blackout system (used for tests)
        self.force_blackout = tf.Variable(False, trainable=False, dtype=tf.bool)
        self.force_p_value = tf.Variable(P_MIN, trainable=False, dtype=tf.float32)
        self.last_p_mean = tf.Variable(0.0, trainable=False)

        # ----- Stage 1
        st = 1 if overlap else 2
        self.s1b1 = GatedResidualBlock(32, 3, stride=st,
                                       blackout_switch=self.force_blackout,
                                       force_p_value_link=self.force_p_value)
        self.s1b2 = GatedResidualBlock(32, 3,
                                       blackout_switch=self.force_blackout,
                                       force_p_value_link=self.force_p_value)
        self.s1_pool = layers.AveragePooling2D(2,2,"same") if overlap else layers.Lambda(lambda z:z)
        self.all_grb_layers += [self.s1b1, self.s1b2]

        # ----- Stage 2
        st = 1 if overlap else 2
        self.s2b1 = GatedResidualBlock(64, 3, stride=st,
                                       blackout_switch=self.force_blackout,
                                       force_p_value_link=self.force_p_value)
        self.s2b2 = GatedResidualBlock(64, 3,
                                       blackout_switch=self.force_blackout,
                                       force_p_value_link=self.force_p_value)
        self.s2_pool = layers.AveragePooling2D(2,2,"same") if overlap else layers.Lambda(lambda z:z)
        self.all_grb_layers += [self.s2b1, self.s2b2]

        # ----- Stage 3
        st = 1 if overlap else 2
        self.s3b1 = GatedResidualBlock(128, 3, stride=st,
                                       blackout_switch=self.force_blackout,
                                       force_p_value_link=self.force_p_value)
        self.s3b2 = GatedResidualBlock(128, 3,
                                       blackout_switch=self.force_blackout,
                                       force_p_value_link=self.force_p_value)
        self.s3_pool = layers.AveragePooling2D(2,2,"same") if overlap else layers.Lambda(lambda z:z)
        self.all_grb_layers += [self.s3b1, self.s3b2]

        # Head
        self.head = keras.Sequential([
            layers.GlobalAveragePooling2D(),
            layers.Dense(128, activation="relu"),
            layers.Dense(num_classes)
        ])

    def call(self, x, training=None):

        x = self.stem(x)
        x = self.bn0(x, training=training)
        x = tf.nn.relu(x)

        # Stage 1
        x = self.s1b1(x, training=training)
        x = self.s1b2(x, training=training)
        x = self.s1_pool(x)

        # Stage 2
        x = self.s2b1(x, training=training)
        x = self.s2b2(x, training=training)
        x = self.s2_pool(x)

        # Stage 3
        x = self.s3b1(x, training=training)
        x = self.s3b2(x, training=training)
        x = self.s3_pool(x)

        # Head
        logits = self.head(x, training=training)

        # Update global p-mean
        p_list = [tf.reduce_mean(l.last_p_map_raw) for l in self.all_grb_layers if l.last_p_map_raw is not None]
        if p_list:
            self.last_p_mean.assign(tf.reduce_mean(tf.stack(p_list)))

        return logits

    def get_raw_p_maps(self, x):
        """
        Returns raw p-maps from each GRB.
        Used in the Visual Proof.
        """
        self.force_blackout.assign(False)
        _ = self(x, training=False)
        out = []
        for lyr in self.all_grb_layers:
            out.append(None if lyr.last_p_map_raw is None else lyr.last_p_map_raw.numpy())
        return out



# ---------------------------------------------------------
# 4) Build Model
# ---------------------------------------------------------
def build_model():
    model = MTMAModel(num_classes=NUM_CLASSES, overlap=OVERLAP)
    dummy = tf.ones((1, IMG_SIZE, IMG_SIZE, 1))
    model(dummy)   # Build model
    model.summary()
    return model

model = build_model()



# ---------------------------------------------------------
# 5) Loss, Optimizer, Metrics
# ---------------------------------------------------------
task_loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True)
optimizer = keras.optimizers.Adam(1e-3)

train_acc = keras.metrics.SparseCategoricalAccuracy()
val_acc   = keras.metrics.SparseCategoricalAccuracy()



# ---------------------------------------------------------
# 6) Training & Validation Steps
# ---------------------------------------------------------
@tf.function
def train_step(x, y):
    model.force_blackout.assign(False)
    with tf.GradientTape() as tape:
        logits = model(x, training=True)
        L_task = task_loss(y, logits)
        L_reg = tf.add_n(model.losses) if model.losses else 0.0
        L_total = L_task + L_reg
    grads = tape.gradient(L_total, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
    train_acc.update_state(y, logits)
    return L_total, L_reg

@tf.function
def val_step(x, y):
    logits = model(x, training=False)
    L_task = task_loss(y, logits)
    val_acc.update_state(y, logits)
    return L_task



# ---------------------------------------------------------
# 7) Training Loop
# ---------------------------------------------------------
print("\n===============================================")
print("== MTMA v0.6.1 Training (Noise-Augmented) ==")
print("===============================================\n")

history = {
    "loss": [], "val_loss": [],
    "acc": [], "val_acc": [],
    "p_mean": []
}

for epoch in range(1, EPOCHS+1):
    train_acc.reset_state()
    val_acc.reset_state()

    # Training on noisy data
    losses = []
    for xb, yb in train_ds:
        L_total, L_reg = train_step(xb, yb)
        losses.append(L_total.numpy())

    # Validation on clean data
    val_losses = []
    for xb, yb in test_ds_clean:
        L = val_step(xb, yb)
        val_losses.append(L.numpy())

    # Store metrics
    history["loss"].append(float(np.mean(losses)))
    history["val_loss"].append(float(np.mean(val_losses)))
    history["acc"].append(float(train_acc.result().numpy()))
    history["val_acc"].append(float(val_acc.result().numpy()))
    history["p_mean"].append(float(model.last_p_mean.numpy()))

    print(f"[Epoch {epoch}] "
          f"loss={history['loss'][-1]:.4f}  "
          f"val_loss={history['val_loss'][-1]:.4f}  "
          f"acc={history['acc'][-1]:.4f}  "
          f"val_acc={history['val_acc'][-1]:.4f}  "
          f"p_mean={history['p_mean'][-1]:.3f}")

learned_optimum_p = history["p_mean"][-1]
learned_optimum_acc = history["val_acc"][-1]



# ---------------------------------------------------------
# 8) Plot Training Curves
# ---------------------------------------------------------
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.plot(history["loss"], label="train_loss (noisy)")
plt.plot(history["val_loss"], label="val_loss (clean)")
plt.legend(); plt.title("Loss")

plt.subplot(1,3,2)
plt.plot(history["acc"], label="train_acc (noisy)")
plt.plot(history["val_acc"], label="val_acc (clean)")
plt.legend(); plt.title("Accuracy")

plt.subplot(1,3,3)
plt.plot(history["p_mean"], label="p_mean")
plt.axhline(BALANCE_TARGET, color="red", linestyle="--", label="Target (0.5)")
plt.legend(); plt.title("Mean(p) Evolution")

plt.tight_layout()
plt.show()



# ---------------------------------------------------------
# 9) TEST 1: Efficiency–Accuracy Sweep Test
# ---------------------------------------------------------
print("\n==============================")
print("== TEST 1: SWEEP TEST ==")
print("==============================\n")

p_levels = np.linspace(P_MIN, 1.0, 15)
sweep_accuracies = []

# Force blackout override (p = constant)
model.force_blackout.assign(True)

for p_value in p_levels:
    model.force_p_value.assign(p_value)
    val_acc.reset_state()

    for xb, yb in test_ds_clean:
        val_step(xb, yb)

    acc = val_acc.result().numpy()
    sweep_accuracies.append(acc)
    print(f"Forced p={p_value:.3f} → Accuracy={acc*100:.2f}%")

model.force_blackout.assign(False)

# Plot Sweep
plt.figure(figsize=(10,6))
plt.plot(p_levels, sweep_accuracies, marker="o", linestyle="--")
plt.axvline(learned_optimum_p, color="red", linestyle=":", label="Learned p")
plt.axhline(learned_optimum_acc, color="red", linestyle=":")
plt.plot(learned_optimum_p, learned_optimum_acc, "r*", markersize=15)
plt.xlabel("Forced Global p")
plt.ylabel("Accuracy")
plt.title("Efficiency vs Accuracy Sweep Test")
plt.grid(True)
plt.legend()
plt.show()



# ---------------------------------------------------------
# 10) TEST 2: Noise Robustness Test
# ---------------------------------------------------------
print("\n==============================")
print("== TEST 2: NOISE ROBUSTNESS ==")
print("==============================\n")

# Dense mode: p = 1.0
model.force_blackout.assign(True)
model.force_p_value.assign(1.0)

val_acc.reset_state()
for xb, yb in test_ds_noisy:
    val_step(xb, yb)
dense_noisy_acc = val_acc.result().numpy()
print(f"Dense (p=1.0) Accuracy on Noisy Data: {dense_noisy_acc*100:.2f}%")

# M-TriX true mode (learned p)
model.force_blackout.assign(False)

val_acc.reset_state()
for xb, yb in test_ds_noisy:
    val_step(xb, yb)
mtrix_noisy_acc = val_acc.result().numpy()
print(f"M-TriX Accuracy on Noisy Data: {mtrix_noisy_acc*100:.2f}%")

print("\nConclusion:")
if mtrix_noisy_acc > dense_noisy_acc * 1.2:
    print("SUCCESS: M-TriX is significantly more robust to noise.")
else:
    print("FAILURE: Robustness not significantly higher.")



# ---------------------------------------------------------
# 11) TEST 3: Visual Proof — Clean p-Map
# ---------------------------------------------------------
print("\n==============================")
print("== TEST 3: VISUAL PROOF ==")
print("==============================\n")

idx = np.random.choice(len(x_test))
orig = x_test[idx]
noisy = add_noise(np.expand_dims(orig, 0), NOISE_FACTOR)[0]

p_maps = model.get_raw_p_maps(noisy[None])
pmap = p_maps[0].squeeze()

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(orig.squeeze(), cmap="gray")
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(noisy.squeeze(), cmap="gray")
plt.title("Noisy Input (30%)")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(pmap, cmap="viridis", vmin=0, vmax=1)
plt.title("M-TriX p-Map (Clean Filter)")
plt.axis("off")
plt.colorbar()

plt.tight_layout()
plt.show()
```
